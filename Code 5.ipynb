### 3번)

```python
class Counter:
    def __init__(self, count, step=1):
        self.count = count
        self.step = step
        
    def __str__(self):
        return "[Count(step: {0})]{1}".format(self.step, self.count)
    
    def incr(self):
        self.count += self.step
    
    def __call__(self):
        return self.incr()
        
    def __add__(self, n):
        newCount = self.count + n
        return Counter(newCount, self.step)
        
    def __sub__(self, n):
        newCount = self.count - n
        return Counter(newCount, self.step)
    
    def __eq__(self, n):
        return self.count == n
    
    def __lt__(self, n):
        return self.count < n
    
    def __gt__(self, n):
        return self.count > n
```

```python
a = Counter(10)
a.incr()
a = a + 5
print(a)
a()
print(a)

print(a > 10)
print(a < 10)
print(a == 10)
```

#### : __init__() 함수는 생성자로 클래스의 인스턴스를 생성 시 호출된다. 현재 MySet 클래스의 __init__ 함수는 함수를 부른 인스턴스와 리스트를 인자로 받아서 self에 요소 추가 함수인 append를 이용하여  하나씩 추가를 해주고, static 메소드인 eliminate_duplicate 중복된 요소 제거 함수를 통해 집합 자료형을 구성하고 있다. 

#### __str__() 함수는 인스턴스 자체를 출력해주는데 MySet의 __str__ 함수는 result 문자열을 만들어 원하는 형식을 지정하여 self에 담겨있는 요소를 출력해준다. 
#### elimicate_duplicate()  함수는 static 함수로 인스턴스를 만들지 않아도 클래스 이름으로 호출 가능하다. MySet 클래스에서 elimicate_duplicate 함수는 주어진 리스트를 받아 중복된 요소를 제거한 새로운 리스트를 만들고, 이 리스트를 다시 인자로 받은 리스트에 요소 하나씩 append 해주는 함수이다.

### 5번)

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    def __or__(self, other): # 합집합
        s = [] # 새로운 리스트 생성
        for e in self: # 새로운 리스트 s에 첫번째 집합 모든 요소 추가
            s.append(e)
        for e in other: # 새로운 리스트 s에 두번째 집합 모든 요소 추가
            s.append(e)
        MySet.eliminate_duplicate(s) # 중복 요소 제거
        return MySet(s) # MySet 객체로 s return
            
    def __and__(self, other): # 교집합
        s = [] # 새로운 리스트 생성
        for e in self: # 첫번째 집합 요소 중 other에 있는 요소만 s에 삽입
            if e in other:
                s.append(e)
        return MySet(s) # MySet 객체로 s return
    
    def __sub__(self, other): # 차집합
        s = [] # 새로운 리스트 생성
        for e in self: # 새로운 리스트 s에 첫번째 집합 모든 요소 추가
            s.append(e)
        for e in s: # s 요소 중 other에 존재하는 요소는 제거
            if e in other:
                s.remove(e)
        return MySet(s) # MySet 객체로 s return
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    u = s | t # 합집합 test
    print(u)
    u = s & t # 교집합 test
    print(u)
    u = t - s # 차집합 test
    print(u)
```

```python
import string
import collections
import requests

class SearchEngine(list):

    def __init__(self,*args):
        index = 0
        for e in args:
            self.append(e)
            
    
    def addUrl(self, other):
        self.append(other)
        
    def removeUrl(self, other):
        self.remove(other)
    
    def listUrls(self):
        for e in self:
            print(e)
            
    @staticmethod
    def count_word(s): # Homework 4까지 수행한 함수
        words = list()
        new = str()
        i = 0
        j = i+1
    
        while(s.find('<',i)!=-1 and j<len(s)):
            if(s.startswith('<style>',j)==True):
                while(s.startswith('</style>',j)==False):
                    j+=1
                i=j+8
                j=i+1
            elif(s.startswith('<script>',j)==True):
                while(s.startswith('</script>',j)==False):
                    j+=1
                i=j+9
                j=i+1
            elif(s[i]=='<'):
                while(s[j]!='>'):
                    if(j>=len(s)):
                        break
                    j+=1
                i=j+1
                j=i+1
            else:
                while(s[j]!='<'):
                    j+=1
                    if(j>=len(s)):
                        break
                if(j==len(s)):
                    new=s[i:]
                else:
                    new=s[i:j]
                words+=new.split()
                i=j
                j=i+1
            
        for p in string.punctuation:
            words = [word.strip(p) for word in words]
    
    
        words = [k for k in words if k not in '']
    
        words = [k for k in words if k not in stop_kor]
        words = [k for k in words if k not in stop_eng]
        
        ret = collections.Counter(words)

        return ret
    
    
    def getWordsFrequency(self, reverse=False):
        temp = []
        for i in self:
            r = requests.get(i)
            source = r.text
            temp.append(source)
        
        temp = "".join(temp)
        
        ret = {}
        ret = self.count_word(temp)
        
        return dict(ret)
    
    def getMaxFreqencyWords(self):
        temp = self.getWordsFrequency()
        maxValue = max(temp.values())
        
        maxDict = {}
        for key, value in temp.items():
            if(value == maxValue):
                maxDict[key] = value
        return maxDict
    
    def searchUrlByWord(self, find):
        d2 = {}
        for i, v in enumerate(self):
            r = requests.get(v)
            source = r.text 
            d1 = {}
            d1 = dict(self.count_word(source))
            d2[i] = d1.get(find, 0) # default값 0으로 설정

        temp = sorted(d2.items(), reverse=True, key=lambda item: item[1])

        for i, j in temp:
            print(self[i])
    
# test
a = SearchEngine('https://www.10000recipe.com/', 'http://www.jbnews.com/', 'https://www.koreatech.ac.kr')
a.addUrl('https://www.naver.com')
a.removeUrl('https://www.koreatech.ac.kr')

a.getMaxFreqencyWords()
a.searchUrlByWord("지금")
a.searchUrlByWord("한국")
a.searchUrlByWord("뉴스")
```
