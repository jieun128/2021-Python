### 5번)

```python
# 1) addall(list) 함수 

# 1-1) for ~ in 리터널
def addall1(list):
    sum = 0
    for i in list:
        sum += i
    return sum

print(addall1([1]))

# 1-2) 리스트 내포 방식
def addall2(list):
    return sum([x for x in list])

print(addall2([1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

```python
# 2) addallodd(list) 함수

# 2-1) for ~ in 리터널
def addall1(list):
    sum = 0
    for i in list:
        if(i%2==1):
            sum += i
    return sum

print(addall1([1]))

# 2-2) 리스트 내포 방식
def addall2(list):
    return sum([x for x in list if x%2==1])

print(addall2([1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

### 8번)

```python
def isStrange(num):
    length = len(str(num))
    
    if num < 0:
        return False
    if length == 1:
        return True
    elif num % length == 0:
        return True
    elif isStrange(num / length):
        return True
    
    return False

print(isStrange(0))
```

### 9번)

```python
def solve(a):
    result = []
    max = 0

    for i in range(len(a)):
        max += a[i]
        
        if(i==len(a)-1):
            result.append(max)
            break
        else:
            if(isOk(max, a[i+1:])==True):
                result.append(max)               
    return result


def isOk(max, b):
    curr = 0   
    
    while(curr < len(b)):
        temp = b[curr]
        if b[curr] == max:
            curr += 1
            continue
        elif(b[curr] < max):
            plus = 1
            if plus < len(b)-curr:
                temp += b[curr+plus]
            else:
                return False
            while(plus < len(b)-curr):
                if(temp > max):
                    return False
                elif temp==max:
                    curr += plus +1
                    break
                else:
                    plus += 1
                    if plus < len(b)-curr:
                        temp += b[curr+plus]
                    else:
                        return False
        else:
            return False
        
    return True
    
# 테스트 
print(solve([1, 2, 1, 1, 1, 2, 1, 3]))
print(solve([1, 1, 1, 1]))
```

### 10번) [Incremental Project 문제]

```python
import requests

req1 = requests.get('http://cse.koreatech.ac.kr')
req2 = requests.get('https://www.koreatech.ac.kr')
req3 = requests.get('http://www.naver.com')
req4 = requests.get('http://www.daum.net')
req5 = requests.get('http://www.nytimes.com')

source1 = req1.text
source2 = req2.text
source3 = req3.text
source4 = req4.text
source5 = req5.text
```

```python
import string
import collections

def count_word(s):
    words = list()
    new = str()
    i = 0
    j = i+1
    
    while(s.find('<',i)!=-1 and j<len(s)):
        if(s.startswith('<style>',j)==True):
            while(s.startswith('</style>',j)==False):
                j+=1
            i=j+8
            j=i+1
        elif(s.startswith('<script>',j)==True):
            while(s.startswith('</script>',j)==False):
                j+=1
            i=j+9
            j=i+1
        elif(s[i]=='<'):
            while(s[j]!='>'):
                if(j>=len(s)):
                    break
                j+=1
            i=j+1
            j=i+1
        else:
            while(s[j]!='<'):
                j+=1
                if(j>=len(s)):
                    break
            if(j==len(s)):
                new=s[i:]
            else:
                new=s[i:j]
            words+=new.split()
            i=j
            j=i+1
            
    for p in string.punctuation:
        words = [word.strip(p) for word in words]
            
    words = [k for k in words if k not in '']
    
    ret = collections.Counter(words)

    return ret

# 테스트
print(count_word(source1))
print(count_word(source2))
print(count_word(source3))
print(count_word(source4))
print(count_word(source5))
```
