### 2번)

```python
def div_path(s):
    div = list()
    count = s.count('/')
    num = 0

    for i in range(0, len(s)):
        if(s[i]=='/'):
            num+=1
            if(num==count):
                div.append(s[0:i])
                div.append(s[i+1:])      
    return div

# 결과 확인
print(div_path('/usr/local/bin/python'))
print(div_path('/home/chulsoo/test.txt'))
```

### 3번)

```python
def list_union(lista, listb):
    tmp = list()
    tmp = lista
    
    for i in listb:
        if i not in tmp:
            tmp.append(i)
            
    tmp.sort()
    return tmp

# 결과 확인
print(list_union([1, 2, 3], [1, 2, 4]))
print(list_union([-10, -5, 0, -1], [100, 9, 0, 9]))
print(list_union([0, 1, 2], [0, 1, 2]))
```

### 4번)

```python
def list_intersection(lista, listb):
    tmp = list()
    
    for i in listb:
        if i in lista:
            tmp.append(i)
            
    tmp.sort()
    return tmp

# 결과 확인
print(list_intersection([1, 2, 3], [1, 2, 4]))
print(list_intersection([-10, -5, 0, -1], [100, 9, 0, 9]))
print(list_intersection([0, 1, 2], [0, 1, 2]))
```

```python
def print_primenumber(a, b):
    prime = list()
    count = 0
    
    for i in range(a, b+1):
        for j in range(1, i+1):
            if(i%j==0):
                count+=1
        if(count==2):
            prime.append(i)
        count=0
        
    print('{} 이상 {} 이하 정수 중 소수 리스트: {}'.format(a, b, prime))
                

# 결과 확인
print_primenumber(2, 10)
```

### 6번) [Incremental Project 문제]

```python
import requests

req1 = requests.get('http://cse.koreatech.ac.kr')
req2 = requests.get('http://link.koreatech.ac.kr/')

source1 = req1.text
source2 = req2.text
```

```python
def count_word(s):
    word = list()
    new = str()
    i = 0
    j = i+1
    
    while(s.find('<',i)!=-1 and j<len(s)):
        if(s.startswith('<style>',j)==True):
            while(s.startswith('</style>',j)==False):
                j+=1
            i=j+8
            j=i+1
        elif(s.startswith('<script>',j)==True):
            while(s.startswith('</script>',j)==False):
                j+=1
            i=j+9
            j=i+1
        elif(s[i]=='<'):
            while(s[j]!='>'):
                if(j>=len(s)):
                    break
                j+=1
            i=j+1
            j=i+1
        else:
            while(s[j]!='<'):
                j+=1
                if(j>=len(s)):
                    break
            if(j==len(s)):
                new=s[i:]
            else:
                new=s[i:j]
            word+=new.split()
            i=j
            j=i+1
    return len(word)
            

# 테스트
print(count_word(source1))
print(count_word(source2))
```
