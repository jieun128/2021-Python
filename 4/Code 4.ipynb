### 3)

```python
def frange(*args):
    if(len(args)==1):
        start = 0.0
        end = args[0]
        step = 0.1
    elif(len(args)==2):
        start, end = args
        step = 0.1
    else:
        start, end, step = args
        
    temp = start
    ret = []
    
    while(True):
        ret.append(temp)
        temp+=step
        if(temp>=end):
            break
            
    return ret

print(frange(0.5))
print(frange(1.0, 2.0))
print(frange(2.2, 4.0, 0.5))
```

```python
import string
import collections


def count_word(s):
    words = list()
    new = str()
    i = 0
    j = i+1
    
    while(s.find('<',i)!=-1 and j<len(s)):
        if(s.startswith('<style>',j)==True):
            while(s.startswith('</style>',j)==False):
                j+=1
            i=j+8
            j=i+1
        elif(s.startswith('<script>',j)==True):
            while(s.startswith('</script>',j)==False):
                j+=1
            i=j+9
            j=i+1
        elif(s[i]=='<'):
            while(s[j]!='>'):
                if(j>=len(s)):
                    break
                j+=1
            i=j+1
            j=i+1
        else:
            while(s[j]!='<'):
                j+=1
                if(j>=len(s)):
                    break
            if(j==len(s)):
                new=s[i:]
            else:
                new=s[i:j]
            words+=new.split()
            i=j
            j=i+1
            
    for p in string.punctuation:
        words = [word.strip(p) for word in words]
    
    
    words = [k for k in words if k not in '']
    
    words = [k for k in words if k not in stop_kor]
    words = [k for k in words if k not in stop_eng]
    
    ret = collections.Counter(words)

    return ret


c1 = count_word(source1)
c2 = count_word(source2)
c3 = count_word(source3)
c4 = count_word(source4)
c5 = count_word(source5)

# print(c1)
# print(c2)
# print(c3)
# print(c4)
# print(c5)
```
