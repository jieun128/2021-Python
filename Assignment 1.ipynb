### 10번)

```python
def appendDelete(s, t, k):
    
    s = list(s)
    t = list(t)
    count = 0
    i = 0

    if len(s) >= len(t):
        while i < len(t):
            if s[i] != t[i]:
                break
            i += 1
        # 삭제 필요 만큼 count up
        count += len(s) - i
        # 추가 필요 만큼 count up
        count += len(t) - i
        del s[i:len(s)]
    else:
        while i < len(s):
            if s[i] != t[i]:
                break
            i += 1
        # 삭제 필요 만큼 count up
        count += len(s) - i
        # 추가 필요 만큼 count up
        count += len(t) - i
        del s[i:len(s)]
        
    # if문 통해서 반환 경우 나누기
    if count == k:
        return 'Yes'
    elif count > k:
        return 'No'
    else: # count가 k보다 작을 경우, s 모두 삭제하고 empty string에서 삭제 연산 진행할 수 있는지 파악
        count += (len(s) * 2)
        if count <= k:
            return 'Yes'
        else:
            return 'No'

# 결과 확인        
print(appendDelete('hackerhappy', 'hackerrank', 9))
print(appendDelete('aba', 'aba', 7))
print(appendDelete('ashley', 'ash', 2))

# [풀이 과정]
# 첫번째 if문에서는 s, t 인덱스 0부터 비교 시작하여 다른 곳 있는 순간 catch하여 그것을 기점으로
# 서로 다른 부분 고치는 count만 세고, 마지막에 del 이용하여 s에서 필요 없는 부분 실제 삭제.

# 두번째 if문 (결과 출력)에서 count > k일 경우 무조건 'No' return
# count == k일 경우 무조건 'Yes' return
# count < k일 경우 남은 s의 요소 개수 * 2 (추가/삭제 연산 진행하는 최소 count) 더해준 count가 k 이하인지 보기.
# 이것이 k 이하라면 empty string에서 삭제 연산 시 홀수번(1번씩) 계산 가능하기에 무조건 가능. ('Yes'  return)
```
